```python
train_df = pd.read_csv("train-1.csv")
test_df = pd.read_csv("test.csv")
```

```python
train_df['Data'] = 'Train'
test_df['Data'] = 'Test'
data_df = pd.concat([train_df, test_df], axis=0).reset_index(drop=True)
data_df['subject'] = '#' + data_df['subject'].astype(str)

train_df.shape, test_df.shape
```

```python
def basic_details(df):
    b = pd.DataFrame()
    b['Missing value'] = df.isnull().sum()
    b['N unique value'] = df.nunique()
    b['dtype'] = df.dtypes
    return b

basic_details(data_df)
```

```python
activity = data_df['Activity']
label_counts = activity.value_counts()

plt.figure(figsize=(12,8))
plt.bar(label_counts.index, label_counts)
```

```python
Data = data_df['Data']
Subject = data_df['subject']
train_df = data_df.copy()
train_df = train_df.drop(['Data','subject','Activity'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
slc = StandardScaler()
label_encoder = LabelEncoder()
train_df = slc.fit_transform(train_df)
activity = label_encoder.fit_transform(activity)

from sklearn.decomposition import PCA
pca = PCA(n_components=0.9, random_state=0)
train_df = pca.fit_transform(train_df)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_df, activity, test_size=0.2, random_state=0)
```

```python
num_folds = 10
seed = 0
scoring = 'accuracy'
results = {}
accuracy = {}
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.model_selection import KFold, cross_val_score
model = KNeighborsClassifier(algorithm='auto', n_neighbors=8, p=1, weights='distance')

_ = cross_val_score(model, X_train, y_train, cv=10, scoring=scoring)
results['GScv'] = (_.mean(), _.std())

model.fit(X_train, y_train)
y_predict = model.predict(X_test)

accuracy['GScv'] = accuracy_score(y_test, y_predict)

print(classification_report(y_test, y_predict))

cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
from xgboost import XGBClassifier
model_XGB = XGBClassifier()
model_XGB.fit(
    X_train,
    y_train
)
```

```python
y_preds = model_XGB.predict(X_test)
print(y_preds[:10])
print(y_preds.shape)
```

```python
accuracy = accuracy_score(y_test, y_preds)
print('Accuracy: %.2f%%' % (accuracy * 100.0))
```
